```python
import sys
sys.path.append("../")

from ai_hub.modules.tts.hf_tts import HuggingFaceTTS
import os
from dotenv import load_dotenv

load_dotenv(override=True)

tts = HuggingFaceTTS(url=os.getenv("HF_TTS_URL"), token=os.getenv("HF_TOKEN"))
```

```python
response_dict = tts.synthesize("आदरणीय उपस्थित मंडळी, माझ्या भगिनींनो आणि बंधूंनो")
print (response_dict)
```

```python
audio_int16 = response_dict["audio"]
sampling_rate = response_dict["sampling_rate"]
```

```python
# play the audio
import IPython.display as ipd
ipd.Audio(audio_int16, rate=sampling_rate)
```
